```python
class Bank():

    def __init__(self):
        self.__account = {}
        self.__balance = {}

    #register method
    def register_user (self, username, password):
        if(self.__account.get(username, "not exist") == "not exist"):
            self.__account[username] = password
            self.__balance[username] = 0
            return True
        else:
             print();print("Username already exist")
             return False

    #Login method
    def login_user(self, username, password):
        if(self.__account.get(username) == password):
            print("Login Successful")
            return True
        else:
             print();print("Login Failed")
             return False

    #Cheak Balance method
    def cheak_balance(self, username):
        if self.__balance.get(username, "not exist") != "not exist":
            print();print(self.__balance[username])
        else:
            print();print("User Not Found")

    #Deposit method
    def deposit(self, username, amount):
        if self.__balance.get(username, "not exist") != "not exist" and amount > 0:
            self.__balance[username] += amount
            print();print("Deposit Successful")
        else:
            print();print(("Invalid amount or User not found" ))
        
    #withdraw method
    def withdraw(self, username, amount):
        if self.__balance.get(username, "not exist") != "not exist" and amount > 0 and amount <= self.__balance.get(username):
            self.__balance[username] -= amount
            print();print("Withdraw Successful")
        else:
            print();print(("Invalid amount or User not found" ))
```
